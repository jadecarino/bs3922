```python
# import matplotlib.pyplot as plt
# Plot a bar graph
# average_salary_by_ed_level.plot(kind='bar', color='skyblue', alpha=0.7)
# # Add labels and title
# plt.xlabel('Highest Education Level attained')
# plt.ylabel('Average Salary ($)')
# plt.title('Average Salary by Education Level')
# # Show the plot
# plt.show()

import seaborn as sns
from matplotlib import pyplot

# color_palette = sns.color_palette("tab10")

avgsalary_edlevel_2019["ds"] = "2019"
avgsalary_edlevel_2020["ds"] = "2020"
avgsalary_edlevel_2021["ds"] = "2021"
avgsalary_edlevel_2022["ds"] = "2022"
avgsalary_edlevel_2023["ds"] = "2023"

dss = pd.concat([avgsalary_edlevel_2019, avgsalary_edlevel_2020, avgsalary_edlevel_2021, avgsalary_edlevel_2022, avgsalary_edlevel_2023])
bar_plot_1 = sns.barplot(x="EdLevel", y="AverageSalary", hue="ds", data=dss, palette=sns.color_palette("Set2"))
bar_plot_1.set_xticklabels(pyplot.xticks()[1], rotation=45, horizontalalignment="right")
bar_plot_1.set_xlabel("Highest Education Level Attained")
bar_plot_1.set_ylabel("Average Yearly Salary ($)")
bar_plot_1.set_title("Average yearly salary of a developer based on the highest level of education attained.")

bar_plot_1.get_figure().savefig("figures/avgsalary_edlevel_barchart.png", bbox_inches = "tight")
```
