```python
print(df_2023.groupby(["RemoteWork"])["ConvertedCompYearly"].mean())
```

```python
avgsalary_remotework_2023 = pd.DataFrame({"RemoteWork": ["Fully in-person", "Hybrid", "Fully remote"],
                    "AverageSalary": [64265, 86496, 100977]})
```

```python
import seaborn as sns
from matplotlib import pyplot

# color_palette = sns.color_palette("tab10")

avgsalary_remotework_2019["ds"] = "2019"
avgsalary_remotework_2022["ds"] = "2022"
avgsalary_remotework_2023["ds"] = "2023"

dss = pd.concat([avgsalary_remotework_2019, avgsalary_remotework_2022, avgsalary_remotework_2023])
bar_plot_1 = sns.barplot(x="RemoteWork", y="AverageSalary", hue="ds", data=dss, palette=sns.color_palette("Set2"))
bar_plot_1.set_xticklabels(pyplot.xticks()[1], rotation=45, horizontalalignment="right")
bar_plot_1.set_xlabel("Working Situation")
bar_plot_1.set_ylabel("Average Yearly Salary ($)")
bar_plot_1.set_title("Average yearly salary of a developer based on their working situation.")

bar_plot_1.get_figure().savefig("figures/avgsalary_remotework_barchart.png", bbox_inches = "tight")
```
